# Filtrations
`sfsutils` also offers a number of filtrations which can be accessed immediately while parsing. Alternatively, to filter a VCF file directly, use the {class}`~sfsutils.filtration.Filterer` class. Some useful filtrations include {class}`~sfsutils.filtration.DeviantOutgroupFiltration`, {class}`~sfsutils.filtration.CodingSequenceFiltration`, and {class}`~sfsutils.filtration.BiasedGCConversionFiltration`. For a complete list of available filtrations, refer to the API reference.

```python
import sfsutils as sf
```

```python
from sfsutils.settings import Settings

Settings.disable_pbar = True
```

```python
# example for filtration
f = sf.Filterer(
    vcf="resources/genome/betula/biallelic.subset.10000.vcf.gz",
    filtrations=[sf.BiasedGCConversionFiltration()],
    output="genome.gc.vcf.gz"
)

f.filter()
```

Note that all components can easily be customized by extending the corresponding base class.
