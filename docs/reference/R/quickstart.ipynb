# Quickstart
Install `sfsutils` from PyPI (VCF parsing requires the optional `vcf` extra):

```bash

```

Parse a site-frequency spectrum from a polarised VCF and plot it. The example VCF for `Betula spp.` carries ancestral alleles in the `AA` info field, which the {class}`~sfsutils.parser.Parser` uses to polarise each site. `n` is the sample size (number of haplotypes) the spectrum is projected to.

````r
setwd("~/PycharmProjects/SFSUtils/")
````

````r
library(sfsutils)

sf <- load_sfsutils()

sf$Settings$disable_pbar <- TRUE
````

````r
vcf <- "resources/genome/betula/biallelic.polarized.subset.10000.vcf.gz"

sfs <- sf$Parser(n = 10, vcf = vcf)$parse()
````

````r
sfs$plot()
````

The {class}`~sfsutils.parser.Parser` polarises sites using the `AA` info tag by default. See the VCF parsing pages for on-the-fly stratification, ancestral-allele and degeneracy annotation, and filtering.
