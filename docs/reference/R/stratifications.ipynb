# Stratifications
We can also use several stratifications in tandem by specifying a list of stratifications. In this example, we stratify the SFS by degeneracy as well as ancestral base. See the {mod}`~sfsutils.parser` module for a complete list of available stratifications.

```r
setwd("~/PycharmProjects/SFSUtils/")
```

```r
library(sfsutils)

su <- load_sfsutils()

su$Settings$disable_pbar <- TRUE
```

```r
# instantiate parser
p <- su$Parser(
  n = 10,
  vcf = "resources/genome/betula/biallelic.polarized.subset.10000.vcf.gz",
  fasta = "resources/genome/betula/genome.subset.20.fasta",
  gff = "resources/genome/betula/genome.gff.gz",
  annotations = list(
    su$DegeneracyAnnotation()
  ),
  stratifications = list(
    su$DegeneracyStratification(),
    su$AncestralBaseStratification()
  )
)

# parse SFS
spectra <- p$parse()
```

```r
# visualize SFS
spectra$plot()
```
